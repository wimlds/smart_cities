```python
parksInsp = (pd.read_csv("parksInspections.csv")).dropna(how="all")
```

```python
parksInsp["parkName"] = parksInsp["Unnamed: 0"].apply(lambda x: x.split(',')[1])
parksInsp["parkCode"] = parksInsp["Unnamed: 0"].apply(lambda x: x.split(',')[0])
parksInsp.set_index("parkCode", inplace=True)
parksInsp.rename(columns={"Unnamed: 1":"date"}, inplace=True)
parksInsp.date = pd.to_datetime(parksInsp.date)
```

```python
type(parksInsp.date.values[0])
```

```python
parksInsp.head()
```

```python
parksInspMean = parksInsp.groupby(parksInsp.index).last()
```

```python
parksInspMean.head()
```

```python
parksInspMean.describe()
```

```python
parksInspMean["mean"] = parksInspMean.mean(axis=1)
```

```python
parksInspMean.sort_values("mean")[["mean"]]
```
