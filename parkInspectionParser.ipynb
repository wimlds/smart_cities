```python
import xml.etree.ElementTree as ET
from lxml import etree
import pandas as pd
import sys

mytags1 = ["park","ParkId","SiteName","Inspections"]

mytags2 = ["OverallCondition","OverallCleanliness","Cleanliness","Glass","Graffiti","Litter","Weeds","Landscape","Lawns","Trees","Structural","Benches","Fences","PavedSurfaces","PlayEquipment","SafetySurface","Sidewalks"]

parksdf = pd.DataFrame()
parksdict = {}
parkname = ''
ParkId = [''] * 3219
SiteName = [''] * 3219

Date = ''
mystruct = {}
mysubst = {}

def process_subtree(c, path, parksdict):
    #print(ET.tostring(c))
    global parkname
    global Date
    #    print (c.tag, c.text)
    if c.tag == "ParkId":
        parkname = c.text
        
    if c.tag == 'SiteName':
        parkname = parkname + "," + c.text
        parksdict[parkname] = {}
    if c.tag == 'Date':
        Date = c.text
        parksdict[parkname][Date] = {}

    if c.tag in mytags2:
        parksdict[parkname][Date][c.tag] = c.text

    park = parkname
    for ii,c1 in enumerate(c):
        process_subtree(c1, path, parksdict)
    return(parksdict)

nodes = []
import xml.etree.ElementTree as ET
tree = ET.parse("DPR_Inspection_001.xml")
root = tree.getroot()

process_subtree(root, [], parksdict)
print ("\n\n\n\n")

#print([ (i, j, parksdict[i][j]) for i in parksdict.keys()  for j in parksdict[i].keys()])
#sys.exit()

parksdf = pd.DataFrame.from_dict({(i,j):
                                  parksdict[i][j]
                                  for i in parksdict.keys() 
                                  for j in parksdict[i].keys()})

parksdf
```
